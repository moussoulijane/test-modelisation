# Pipeline Forecast Robuste

Ce notebook execute la version robuste du pipeline dans ce dossier. Les chemins sont relatifs au projet copie: `in/` pour l'entree et `out/` pour les sorties.

```python
from time_series_analysis_pipeline_v7 import (
    INPUT_DIR,
    OUTPUT_DIR,
    PipelineConfig,
    ResultsExporter,
    TimeSeriesAnalyzer,
    detect_date_column,
    detect_indicator_columns,
    load_excel_data,
)

print('Input:', INPUT_DIR)
print('Output:', OUTPUT_DIR)
```

```python
df = load_excel_data(INPUT_DIR)
date_col = detect_date_column(df)
indicator_cols = detect_indicator_columns(df, date_col)

print('Date column:', date_col)
print('Numeric indicators:', len(indicator_cols))
df.head()
```

```python
config = PipelineConfig(
    horizon=30,
    holdout_min=30,
    holdout_max=90,
    max_lag=120,
    max_selected_lags=24,
    use_xgboost=True,
)

analyzer = TimeSeriesAnalyzer(df, date_col=date_col, indicator_cols=indicator_cols, config=config)
analyzer.analyze_all()
```

```python
exporter = ResultsExporter(analyzer, OUTPUT_DIR)
excel_path = exporter.create_summary_report()
exporter.create_visualization_dashboard()
exporter.create_individual_charts()

print('Excel:', excel_path)
print('HTML outputs:', OUTPUT_DIR)
```
