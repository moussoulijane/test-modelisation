```python
from time_series_analysis_pipeline_v7 import (
   load_excel_data,
   TimeSeriesAnalyzer,
   ResultsExporter
)
from pathlib import Path
```

```python
df = load_excel_data("C:\work\PROJETS_MODELISATION\Analyses Mr Hamlili\Historique PNB dépots et crédits\Séries_chrono\pl_timeseries_pipeline\in")
date_col = next(
   (c for c in df.columns if 'date' in c.lower()),
   df.columns[0]
)
indicator_cols = [
   c for c in df.columns
   if c != date_col and df[c].dtype in [float, int]
]
analyzer = TimeSeriesAnalyzer(
   df,
   date_col=date_col,
   indicator_cols=indicator_cols
)
analyzer.analyze_all()
```

```python
exporter = ResultsExporter(analyzer, "C:\work\PROJETS_MODELISATION\Analyses Mr Hamlili\Historique PNB dépots et crédits\Séries_chrono\pl_timeseries_pipeline\out")
exporter.create_summary_report()
exporter.create_visualization_dashboard()
exporter.create_individual_charts()
```
